### Task 3.3: find maximum infection rate for all of the countries.

```python
# Initialize an empty list to store max infection rates
max_infection_rates = []

# Loop through each country in the DataFrame
for country in corona_dataset_aggregated.index:
    # Convert the country's data to numeric type to handle any potential string values
    country_data_numeric = pd.to_numeric(corona_dataset_aggregated.loc[country], errors='coerce')

    # Compute the difference between consecutive values and find the maximum
    max_infection_rate = country_data_numeric.diff().max()

    # Append the maximum infection rate to the list
    max_infection_rates.append(max_infection_rate)

# Print the list of maximum infection rates
print(max_infection_rates)
```

### Task 3.4: create a new dataframe with only needed column

```python
# Initialize an empty list to store max infection rates
max_infection_rates = []

# Loop through each country in the DataFrame
for country in corona_dataset_aggregated.index:
    # Convert the country's data to numeric type to handle any potential string values
    country_data_numeric = pd.to_numeric(corona_dataset_aggregated.loc[country], errors='coerce')

    # Compute the difference between consecutive values and find the maximum
    max_infection_rate = country_data_numeric.diff().max()

    # Append the maximum infection rate to the list
    max_infection_rates.append(max_infection_rate)

# Add the list of maximum infection rates as a new column to the DataFrame
corona_dataset_aggregated["max_infection_rates"] = max_infection_rates

# Create a new DataFrame containing only the 'max_infection_rates' column
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rates"])

# Print the new DataFrame
print(corona_data)
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
import seaborn as sns

# Assuming 'data' is your DataFrame containing both "GDP per capita" and "max_infection_rates" columns

# Extracting x and y variables
x = data["GDP per capita"]
y = data["max_infection_rates"]

# Creating the scatter plot
sns.scatterplot(data=data, x=x, y=y)

# Adding labels and title
plt.xlabel("GDP per capita")
plt.ylabel("Max Infection Rate")
plt.title("Scatter Plot of Max Infection Rate vs. GDP per capita")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Assuming 'data' is your DataFrame containing both "GDP per capita" and "max_infection_rates" columns

# Extracting x and y variables
x = data["GDP per capita"]
y = data["max_infection_rates"]

# Creating the scatter plot with log scaling for y
sns.scatterplot(data=data, x=x, y=np.log(y))

# Adding labels and title
plt.xlabel("GDP per capita")
plt.ylabel("Log(Max Infection Rate)")
plt.title("Scatter Plot of Max Infection Rate vs. GDP per capita with Log Scaling")

# Display the plot
plt.show()
```
